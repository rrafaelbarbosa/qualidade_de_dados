```python
# bibliotecas de uso padrão

import pandas as pd
# bibliotecas de analise
import numpy as np
#bibliotecas de função matemática
import matplotlib.pyplot as plt
# biblioteca para gerar gráficos
```

```python
# os dados eles estão sujos (com defeito)
# carregar esses dados
df_dados_oscar = pd.read_csv('/content/exercicio_qualidade_dados_oscar.csv', on_bad_lines='warn')

#tentar ver dados
display(df_dados_oscar.head(20))
```

```python
# ver onde tem dado nulo e seus tipos
df_dados_oscar.info()
```

```python
# criar contagens
print(df_dados_oscar['Ano'].value_counts)
```

```python
# remover tudo que é duplicado
# guardo no sistema o valor atual de linhas
linhas_antes = len(df_dados_oscar)

# remover duplicatas
# para remover eu chamo a base de dados falo que ela recebe ela mesma com o
# atributo drop_duplicate()

df_dados_oscar = df_dados_oscar.drop_duplicates()
display(df_dados_oscar.head(20))
```

```python
# mostrar valores unicos
print(df_dados_oscar['Vencedor'].unique())
```

```python
# vamos converter todos os dados para True ou False
mapa_vencedores = {
    'Yes': True, 'Y': True, 'No': False,
    '1': True, 'Winner!': True, 'Sim': True,
    'TRUE': True, 'N': False, '0': True, 'nan': False
}

# trocando os dados
# chamo a base de dados com a coluna, faço ela receber ela msm
# com a coluna usamos o atributo map para mapear os novos dados
df_dados_oscar['Vencedor'] = df_dados_oscar['Vencedor'].map(mapa_vencedores)
# para eliminar um dado nulo ou vazio ou NaN
# chamo base de dados com a coluna, faço ela receber ela mesma
# com a coluna usamos o atributo fillna('valor desejado')
df_dados_oscar ['Vencedor'] = df_dados_oscar ['Vencedor'].fillna(False)

display(df_dados_oscar.head(20))
```

```python
# corrigir dados escritos errados
# chamo a base de dados com coluna e faço ela receber ela mesma
# e uso o método replace('antes, depois)
df_dados_oscar['Indicado'] = df_dados_oscar['Indicado'].replace({'Oppenheimmer': 'Oppenheimer'})
display(df_dados_oscar.head(2))
```
